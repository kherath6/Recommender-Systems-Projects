### Content-based filtering measures the similarity in the attributes.
**In this case we don't have many attributes to consider:** 
    - star rating can be an atribute however, it is not very descriptive. 
    - review_count may not be the best attribute that describes a restaurent. 
**Hence, we can use the service_categories column to create atributes using the text.** 

### 1. Text Processing

- Remove special characters/ punctuation
- Remove symbols/ numbers  
- Convert to Lowercase  
- Remove stopwords
- Lemmatize

```python
business.service_categories
```

```python
# Unique Categories 
business.service_categories.unique()
```

```python
business.service_categories.nunique()
```

```python
# Define a function to clean text 
clean_spChar = re.compile('[/(){}\[\]\|@,:]')
clean_symbols = re.compile('[^0-9a-z #+_]')
stopWords = stopwords.words('english')
stemmer = nltk.stem.PorterStemmer()

def clean_text(text):
    # Remove white space
    text = text.strip()
    # Convert text to lowercase 
    text = text.lower()
    #Remove the word Restaurant and Restaurants
    text = text.replace('restaurant', '')
    text = text.replace('restaurants', '')
    # Remove special chars and symbols
    text = clean_spChar.sub(' ', text)
    text = clean_symbols.sub(' ', text)
    # Remove stopwords 
    text = ' '.join(word for word in text.split() if word not in stopWords)
    # perform stemming/lemmatization 
    #text = stemmer.stem(text)
    
    return text
```

```python
# Clean, transform the text and save to new column 
business['service_categories_clean'] = business['service_categories'].apply(clean_text)
business.head()
```

### 2. Word Cloud

```python
wordcloud = WordCloud(background_color="white",width=1900, 
                      height=1000).generate(' '.join(business['service_categories_clean'].tolist()))
plt.figure(figsize=(25,15), facecolor='k')
plt.imshow(wordcloud)
plt.show()
```

### 3. TF-IDF Vectorizer

```python
tfIdf = TfidfVectorizer(analyzer='word', ngram_range=(1,3))
tfIdf_matrix = tfIdf.fit_transform(business['service_categories_clean'])
```

```python
print(tfIdf_matrix)
```

### Interpretation 

- Assume general form: (A,B) C

- A: Document index B: Specific word-vector index C: TFIDF score for word B in document A

### 4. Cosine Similarity Matrix 

```python
cos_sim = cosine_similarity(tfIdf_matrix, tfIdf_matrix)
```

```python
pd.DataFrame(cos_sim).head()
```

#### Change the index to **business_name**

```python
business1 = business.copy()
```

```python
business1.set_index('business_name', inplace=True)
```

```python
business1.head()
```

### 5. Create a variable with business_names 

```python
business_names = pd.Series(business1.index)
```

```python
business_names.head(20)
```

### 6. Create Recommendation Model 

```python
# Define a function to recommend a restaurent based on cosine similarity 

def recommender(business_name, cos_sim = cos_sim):
    recommendations = []
    # Index value of the given business name 
    indx = business_names[business_names == business_name].index[0]
    # get the similarity scores and sort values 
    sim_scores = pd.Series(cos_sim[indx]).sort_values(ascending=False)
    # Get the top 10 result indexes
    top_ten_indexes = list(sim_scores[1:11].index)
    # Get business_names 
    for i in top_ten_indexes:
        recommendations.append(list(business1.index)[i])
    return recommendations  
```

### 7. Make Receommendations

```python
# Receommendations for Cafe Lucia

lucia = recommender("Cafe Lucia")
```
